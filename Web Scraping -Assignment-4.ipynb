#### 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
#### Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
#### You need to find following details:
##### A) Rank
##### B) Name
##### C) Artist
#####  D) Upload date
##### E) Views

```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from datetime import datetime
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
driver = webdriver.Chrome(r'C:\Users\udayakiran\Desktop\chromedriver.exe')
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/")

opensite=driver.find_element_by_xpath("//span[@class='plainlinks']/a[1]/i[1]")
opensite.click()

listofmostviewed_link=driver.find_element_by_xpath("//div[@class='mw-search-result-heading']/a[1]/span")
listofmostviewed_link.click()
time.sleep(3)
Rankofvideo,Name,Artist,views,Upload_Date=[],[],[],[],[]
for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody[1]/tr/td[1]")[:30]:
    Rankofvideo.append(i.text.replace(".",""))
#print(Rank)
for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody[1]/tr/td[2]")[:30]:
    Name.append(i.text.replace('"',''))
for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody[1]/tr/td[3]")[:30]:
    Artist.append(i.text)
for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody[1]/tr/td[4]")[:30]:
    views.append(i.text)
for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody[1]/tr/td[5]")[:30]:
    Upload_Date.append(i.text)

List_of_views=pd.DataFrame({"Rank of video":pd.Series(Rankofvideo),
                            "Name Of The Video":pd.Series(Name),
                            "Name Of The Uploader or Artist":pd.Series(Artist),
                            "No. Of Views in Billions":pd.Series(views),
                            "Upload Date":pd.Series(Upload_Date)                           
                           })
List_of_views[:5]
```

#### 2. Scrape the details team India’s international fixtures from bcci.tv.

#### You need to find following details:
#### A) Match title (I.e. 1st ODI)
#### B) Series
#### C) Place
#### D) Date
#### E) Time

```python
driver = webdriver.Chrome(r'C:\Users\udayakiran\Desktop\chromedriver.exe')
driver.get("https://www.bcci.tv/")
on_menu=driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']")
on_menu.click()
fixture_click=driver.find_element_by_xpath("//div[@class='drop-down__options']/ul[1]/li[1]/a[1]")
fixture_click.click()
time.sleep(3)
Date,Match_Title,Series,place,Time =[],[],[],[],[]
for i in driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']"):
    str_splt = (i.text).split()
    str_splt1=str_splt[0]+" "+str_splt[1]+" "+str_splt[2]
    str_splt2=str_splt[3]+" "+str_splt[4]
    Date.append(str_splt1)
    Time.append(str_splt2)
for i in driver.find_elements_by_xpath("//div[@class='fixture__teams']"):
    Series.append(i.text.replace("\n",' '))
for i in driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']"):
    Match_Title.append(i.text.replace("\n",' '))
for i in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span[1]"):
    place.append(i.text.replace("\n",' '))
    
details_fixtures =pd.DataFrame({"Match Titles":pd.Series(Match_Title),
                            "Series":pd.Series(Series),
                            "Place":pd.Series(place),
                            "Date":pd.Series(Date),
                            "Time":pd.Series(Time)                           
                           })
details_fixtures[:5]
```
